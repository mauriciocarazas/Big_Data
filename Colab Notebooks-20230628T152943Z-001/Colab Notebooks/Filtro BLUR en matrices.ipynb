```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

# Generar matrices aleatorias
matrix1 = torch.randn(10, 10)
matrix2 = torch.randn(100, 100)
matrix3 = torch.randn(1000, 1000)

# Definir el filtro de desenfoque
blur_filter = torch.ones(3, 3) / 9.0  # Filtro de promedio

# Aplicar el filtro de desenfoque a cada matriz
blurred_matrix1 = F.conv2d(matrix1.view(1, 1, 10, 10), blur_filter.view(1, 1, 3, 3), padding=1).view(10, 10)
blurred_matrix2 = F.conv2d(matrix2.view(1, 1, 100, 100), blur_filter.view(1, 1, 3, 3), padding=1).view(100, 100)
blurred_matrix3 = F.conv2d(matrix3.view(1, 1, 1000, 1000), blur_filter.view(1, 1, 3, 3), padding=1).view(1000, 1000)

# Imprimir las matrices originales y las matrices desenfocadas
print("Matriz 1:")
print(matrix1)
print("Matriz 1 desenfocada:")
print(blurred_matrix1)

print("Matriz 2:")
print(matrix2)
print("Matriz 2 desenfocada:")
print(blurred_matrix2)

print("Matriz 3:")
print(matrix3)
print("Matriz 3 desenfocada:")
print(blurred_matrix3)

# Visualizar las matrices originales y las matrices desenfocadas
fig, axs = plt.subplots(2, 3, figsize=(12, 8))
axs[0, 0].imshow(matrix1.numpy(), cmap='gray')
axs[0, 0].set_title('Matriz 1')
axs[0, 1].imshow(matrix2.numpy(), cmap='gray')
axs[0, 1].set_title('Matriz 2')
axs[0, 2].imshow(matrix3.numpy(), cmap='gray')
axs[0, 2].set_title('Matriz 3')
axs[1, 0].imshow(blurred_matrix1.numpy(), cmap='gray')
axs[1, 0].set_title('Matriz 1 desenfocada')
axs[1, 1].imshow(blurred_matrix2.numpy(), cmap='gray')
axs[1, 1].set_title('Matriz 2 desenfocada')
axs[1, 2].imshow(blurred_matrix3.numpy(), cmap='gray')
axs[1, 2].set_title('Matriz 3 desenfocada')
plt.show()
```
