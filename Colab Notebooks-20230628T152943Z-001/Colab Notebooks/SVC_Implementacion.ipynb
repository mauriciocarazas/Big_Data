```python
# Importar las bibliotecas necesarias
from sklearn import svm
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Cargar la primera base de datos (iris)
iris = datasets.load_iris()
X_iris = iris.data
y_iris = iris.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

# Crear un clasificador SVM
clf_iris = svm.SVC(kernel='linear')

# Entrenar el clasificador utilizando los datos de entrenamiento
clf_iris.fit(X_train, y_train)

# Realizar predicciones en los datos de prueba
y_pred_iris = clf_iris.predict(X_test)

# Calcular la precisión de las predicciones
accuracy_iris = accuracy_score(y_test, y_pred_iris)
print("Precisión de la base de datos iris:", accuracy_iris)


# Cargar la segunda base de datos (digits)
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits, test_size=0.2, random_state=42)

# Crear un clasificador SVM
clf_digits = svm.SVC(kernel='linear')

# Entrenar el clasificador utilizando los datos de entrenamiento
clf_digits.fit(X_train, y_train)

# Realizar predicciones en los datos de prueba
y_pred_digits = clf_digits.predict(X_test)

# Calcular la precisión de las predicciones
accuracy_digits = accuracy_score(y_test, y_pred_digits)
print("Precisión de la base de datos digits:", accuracy_digits)
```

```python
import numpy as np

class SVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
    
    def fit(self, X, y):
        num_samples, num_features = X.shape

        # Inicializar los parámetros
        self.weights = np.zeros(num_features)
        self.bias = 0

        # Gradiente descendiente para encontrar los parámetros óptimos
        for _ in range(self.num_iterations):
            # Calcular la función de decisión
            linear_model = np.dot(X, self.weights) + self.bias

            # Calcular la función de pérdida (función de costo)
            hinge_loss = np.maximum(0, 1 - y * linear_model)
            loss = 0.5 * np.dot(self.weights, self.weights) + self.lambda_param * np.mean(hinge_loss)

            # Calcular los gradientes
            gradients = self.lambda_param * np.dot(X.T, (linear_model < 1) * y) - self.weights

            # Actualizar los parámetros
            self.weights -= self.learning_rate * gradients
            self.bias -= self.learning_rate * np.mean((linear_model < 1) * y)

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return np.sign(linear_model)

# Ejemplo de uso
X = np.array([[2, 2], [4, 4], [1, 4], [3, 2]])
y = np.array([-1, -1, 1, 1])

svm = SVM()
svm.fit(X, y)

# Datos de prueba
X_test = np.array([[1, 1], [4, 5]])
y_pred = svm.predict(X_test)

print("Predicciones:", y_pred)
```
